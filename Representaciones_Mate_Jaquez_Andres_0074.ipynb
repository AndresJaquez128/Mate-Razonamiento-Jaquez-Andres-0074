Paso 1: Análisis y resoluciónPara resolver esto, sigamos un orden lógico:Definir la variable: Llamaremos $x$ al número desconocido.Plantear la ecuación: El enunciado dice "la suma de un número ($x$) más 9 ($+ 9$) es igual a 21 ($= 21$)".$$x + 9 = 21$$Despejar la variable: Para dejar a la $x$ sola, el 9 que está sumando pasa al otro lado de la igualdad restando.$$x = 21 - 9$$Resolver:$$x = 12$$Resultado: El número es 12.

```python
# Problema 1: Hallar un número x tal que x + 9 = 21
#resultado_total = 21
#numero_conocido = 9
resultado_total = int(input("Ingrese el resultado total: "))
numero_conocido = int(input("Ingrese el número conocido: "))
# Despejamos: x = 21 - 9
incognita = resultado_total - numero_conocido

print(f"El número buscado es: {incognita}")
```

##Problema 2:
Pedro tiene el doble de edad que Ana y suman 36Paso 1: Análisis y resoluciónEste problema requiere relacionar dos variables, pero podemos expresarlo todo en términos de una sola.Definir las variables:Edad de Ana: $x$Edad de Pedro: $2x$ (porque es el doble que la de Ana).Plantear la ecuación: La suma de ambas edades es igual a 36.$$x + 2x = 36$$Simplificar: Sumamos las términos semejantes ($1x + 2x = 3x$).$$3x = 36$$Despejar la variable: El 3 que está multiplicando a la $x$ pasa dividiendo al 36.$$x = \frac{36}{3}$$Resolver:$x = 12$ (Edad de Ana)$2x = 2(12) = 24$ (Edad de Pedro)Resultado: Ana tiene 12 años y Pedro tiene 24 años.

```python
# Problema 2: Edades de Pedro y Ana
#suma_total = 36
suma_total = int(input("Ingrese la suma total: "))
# La ecuación es: x (Ana) + 2x (Pedro) = 36 -> 3x = 36
# Dividimos el total entre 3 para hallar la unidad base (edad de Ana)
edad_ana = suma_total / 3
edad_pedro = edad_ana * 2

print(f"La edad de Ana es: {int(edad_ana)} años")
print(f"La edad de Pedro es: {int(edad_pedro)} años")
```
